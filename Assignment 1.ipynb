5. Data visualization using Scatter Plot

```python
x = dataframe.SepalLengthCm
y = dataframe.SepalWidthCm
plt.scatter(x,y)
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('Scatter Plot of SepalLengthCm Vs SepalWidthCm ')
plt.show()
```

5.(ii)Data visualization using Histogram

```python
plt.hist(dataframe['SepalLengthCm'],bins = 30, edgecolor = 'r', color = 'y')
plt.xlabel('SepalLengthCm')
plt.ylabel('Frequency')
plt.title('Histogram of SepalLengthCm')
plt.show()
```

5.(iii)Data visualization using Distplot

```python
sns.distplot(dataframe.SepalLengthCm,bins=30,color='r')
```

5.(iv)Data visualization using KDEPlot

```python
sns.kdeplot(dataframe.SepalLengthCm,color='g')
```
