```python
data1['average score'] = (data1['math score'] + data1['writing score'] + data1['reading score']) / 3
```

```python
gender = data1['gender'].value_counts()
plt.figure(figsize=(7, 6))
ax = gender.plot(kind='bar', rot=0, color="c")
ax.set_title("Gender Vs Average score", y = 1)
ax.set_xlabel('Gender')
ax.set_ylabel('average score')
ax.set_xticklabels(('Male', 'Female'))
```

Male scores the most on average for math, reading and 
writing based on gender

```python
gender = data1['test preparation course'].value_counts()
plt.figure(figsize=(7, 6))
ax = gender.plot(kind='bar', rot=0, color="c")
ax.set_title("Test preparation Course Vs Average score", y = 1)
ax.set_xlabel('test preparation course')
ax.set_ylabel('average score')
ax.set_xticklabels(('None', 'completed'))
```

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(131)
plt.title('Math Scores')
sns.barplot(x="gender", y="math score", data=data1)
plt.subplot(132)
plt.title('Reading Scores')
sns.barplot(x="gender", y="reading score", data=data1)
plt.subplot(133)
plt.title('Writing Scores')
sns.barplot(x="gender", y="writing score", data=data1)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(131)
plt.title('Math Scores')
sns.barplot(x="test preparation course", y="math score", data=data1)
plt.subplot(132)
plt.title('Reading Scores')
sns.barplot(x="test preparation course", y="reading score", data=data1)
plt.subplot(133)
plt.title('Writing Scores')
sns.barplot(x="test preparation course", y="writing score", data=data1)
plt.show()
```
